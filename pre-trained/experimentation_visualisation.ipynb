```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
columns = ["num_shuffled", "accuracy"]
df = pd.read_csv('../data/accuracy_scores.csv', usecols=columns)
```

```python
plt.plot(df.num_shuffled, df.accuracy)
plt.show()
```

```python
plt.scatter(df.num_shuffled, df.accuracy)

m, b = np.polyfit(df.num_shuffled, df.accuracy, 1)
plt.plot(df.num_shuffled, m*df.num_shuffled+b)
plt.show() 
```
